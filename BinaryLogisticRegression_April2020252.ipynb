# Problem
We need to predict breast cancer based on features like lump size, perimeter, text etc.

## Data Collection

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
# Loading dataset
data = pd.read_csv(r'B:\Training\BIA\Machine Learning\For Students\Day_3\Logistic Regression\breast_cancer_data.csv')
```

```python
# Checking data properties
df = data.copy()
df.head()
```

```python
# To check unique values in target column
df['target'].unique()
```

```python
# Null values
df.isnull().sum()
```

```python
# Duplicates
df[df.duplicated()]
```

```python
# Outliers
df.describe()
```

```python
# Correlation
plt.figure(figsize = (15,6))
sns.heatmap(df.corr(), cmap = 'coolwarm')
plt.show()
```

```python
# Setting Feature and Target
X = df.drop(columns = ['target'], axis = 1)
y = df['target']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
# Data Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Model Building & Training
model = LogisticRegression()

# Training the model
model.fit(X_train_scaled, y_train)
```

```python
# Prediction and Accuracy
y_predict = model.predict(X_test_scaled)
df_predict = pd.DataFrame({'Actua': y_test, 'Predicted': y_predict})
df_predict.head(10)
```

```python
# Accuracy
accuracy = metrics.accuracy_score(y_test, y_predict)
confusion = metrics.confusion_matrix(y_test, y_predict)
f1_score = metrics.f1_score(y_test, y_predict)
classification_report = metrics.classification_report(y_test, y_predict)
print(f'The model accuracy is: {accuracy*100:.2f}%')
print(f'f1 score is: {f1_score:.2f}')
print("The classification report is: \n", "\n", classification_report)
print('The confusion Matrix is: \n', confusion)
```

```python
# ROC - AUC, ROC Curve
y_pred_prob = model.predict_proba(X_test)[:, 1]
roc_auc = metrics.roc_auc_score(y_test, y_pred_prob)
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_prob)
```

```python
# Plotting ROC
plt.figure(figsize = (10,6))
plt.plot(fpr, tpr, color = 'orange', label = f'ROC Curve (AUC = {roc_auc: .2f})')
plt.plot([0,1], [0,1], linestyle = '--')
plt.xlim(0.0, 1.0)
plt.ylim(0.0,1.1)
plt.xlabel("Flase Positive")
plt.ylabel("True Positive")
plt.title("Receiver Operating Characteristics")
plt.legend()
plt.show()
```
